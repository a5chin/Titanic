```python
import pandas as pd
import seaborn as sns; sns.set()
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
def visualize_null(df):
    count_null = df.isnull().sum()
    percent = count_null / len(df) * 100
    data = pd.concat([count_null, percent], axis=1)
    return pd.DataFrame(data={'count_null': count_null, 'percent': percent})
```

```python
visualize_null(train)
```

```python
visualize_null(test)
```
